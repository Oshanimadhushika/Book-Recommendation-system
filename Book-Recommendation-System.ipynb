```python

ratings = ratings[ratings['User-ID'].isin(y)]

print(ratings)

ratings.head(5)
```

```python
ratings.shape
```

```python
books.head(2)
```

```python
ratings_with_book=ratings.merge(books , on="ISBN")

ratings_with_book
```

```python
ratings_with_book.shape
```

```python
num_ratings =ratings_with_book.groupby('title')['rating'].count().reset_index()
num_ratings.head()
```

```python
num_ratings.columns = num_ratings.columns.str.strip()

num_ratings.rename(columns={"rating": "num_of_rating"}, inplace=True)


num_ratings.head(5)
```

```python
final_rating=ratings_with_book.merge(num_ratings , on="title")

final_rating.head(3)
```

```python
final_rating=final_rating[final_rating['num_of_rating']>=50]


final_rating
```

```python
final_rating.sample(10)
```

```python
final_rating.shape
```

```python
 final_rating.drop_duplicates(['User-ID', 'title'], inplace=True)
```

```python
final_rating.shape
```

```python
final_rating
```

```python
book_pivot=final_rating.pivot_table(columns='User-ID',index='title',values='rating')

book_pivot
```

```python
book_pivot.fillna(0,inplace=True)

book_pivot
```

```python
!pip install scipy
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse=csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
!pip install scikit-learn
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distance , suggestion =model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1),n_neighbors=6)
```

```python
def recommend_book(book_name):
    book_id=np.where(book_pivot.index==book_name)[0][0]
    distance , suggestion =model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1),n_neighbors=6)

    for i in range(len(suggestion)):
        books=book_pivot.index[suggestion[i]]
        for j in books:
            print(j)
```

```python
book_name="A Bend in the Road"
recommend_book(book_name)
```
